```python
values=[5,8,9,10,4,7]
col=['b','g','c','m','y','r']
lab=['A','B','C','D','E','F']
exp=[0,0,0,0,0,0.5]
plt.pie(values,colors=col,labels=lab,explode=exp,
        autopct='%.f%%',counterclock=False)
plt.show()
```

```python
country=['Norway','Germany','Canada','United States','Netherlands']
medal=[14,15,11,9,8]
col=['b','g','m','c','y']
exp=[0,0.2,0,0,0]
plt.pie(medal,colors=col,labels=country,explode=exp,autopct='%1.2f%%',shadow=True)
plt.title('Medal Tally',fontsize=15)
plt.show()
```

```python
x=range(1,7)
y=[5,8,9,10,4,7]
col=['b','g','c','m','y','r']
plt.bar(x,y,color=col)
plt.savefig('Bar Chart.png')
plt.show()
```

```python
city=['Delhi','Beijing','Washington','Tokyo','Moscow']
happiness_index=[60,40,70,65,85]
col=['b','g','c','m','y']
plt.xlabel('City')
plt.ylabel('Happiness Index')
plt.title('Happiness Graph', fontsize=20)
plt.grid()
plt.bar(city,happiness_index,color=col)
for x,y in zip(city,happiness_index):
    plt.text(x,y,y)
plt.show()
```
